```python
def SIR_model(y_data):
    ''' Simple SIR model
        S: susceptible population
        I: infected people
        R: recovered people
    '''
    global SIR0, t, N0

    ydata = np.array(y_data)
    t = np.arange(len(ydata))

    N0 = 1000000 #max susceptible population
    I0=ydata[0] #Infected population
    S0=N0-I0    #susceptible population
    R0=0
    SIR0 = (S0,I0,R0)


    popt = [0.2, 0.1]
    fit_odeint(t, *popt)

    popt, pcov = optimize.curve_fit(fit_odeint, t, ydata)
    perr = np.sqrt(np.diag(pcov))

    print('standard deviation errors : ',str(perr), ' start infect:',ydata[0])
    print("Optimal parameters: beta =", popt[0], " and gamma = ", popt[1])

    fitted=fit_odeint(t, *popt)

    return t, fitted
```

```python
def SIR_model_t(SIR,t,beta,gamma):
    ''' Simple SIR model
        S: susceptible population
        t: time step, mandatory for integral.odeint
        I: infected people
        R: recovered people
        beta: 
        
        overall condition is that the sum of changes (differnces) sum up to 0
        dS+dI+dR=0
        S+I+R= N (constant size of population)
    
    '''
    
    S,I,R=SIR
    dS_dt=-beta*S*I/N0          #S*I is the 
    dI_dt=beta*S*I/N0-gamma*I
    dR_dt=gamma*I
    return dS_dt,dI_dt,dR_dt
```

```python
def fit_odeint(x, beta, gamma):
    return integrate.odeint(SIR_model_t, SIR0, t, args=(beta, gamma))[:,1] # we only would like to get dI
```

```python
df_analyse=pd.read_csv('../data/processed/COVID_final_set.csv',sep=';') 
```

```python
fig = go.Figure()
app = dash.Dash()
app.layout = html.Div([

    dcc.Markdown('''

    # Implementation of SIR model

    '''),


    dcc.Dropdown(
        id = 'country_drop_down',
        options=[ {'label': each,'value':each} for each in df_analyse['country'].unique()],
        value= 'Germany', # which is pre-selected
        multi=False),

    dcc.Graph(figure = fig, id = 'SIR_graph')
    ])

def SIR(countries):

    SIR_model()


@app.callback(
    Output('SIR_graph', 'figure'),
    [Input('country_drop_down', 'value')])

def update_figure(country_drop_down):

    traces = []

    df_plot = df_analyse[df_analyse['country'] == country_drop_down]
    df_plot = df_plot[['state', 'country', 'confirmed', 'date']].groupby(['country', 'date']).agg(np.sum).reset_index()
    df_plot.sort_values('date', ascending = True).head()
    df_plot = df_plot.confirmed[35:]

    t, fitted = SIR_model(df_plot)

    traces.append(dict (x = t,
                        y = fitted,
                        mode = 'markers',
                        opacity = 0.9,
                        name = 'SIR-fit')
                  )

    traces.append(dict (x = t,
                        y = df_plot,
                        mode = 'lines',
                        opacity = 0.9,
                        name = 'Original Data')
                  )

    return {
            'data': traces,
            'layout': dict (
                width=1280,
                height=720,
                title = 'Fitted SIR model',

                xaxis= {'title':'Days',
                       'tickangle':-45,
                        'nticks':20,
                        'tickfont':dict(size=14,color="#7f7f7f"),
                      },

                yaxis={'title': "Infected population"}
        )
    }


if __name__ == '__main__':
    app.run_server(debug = True, use_reloader = False)

```
